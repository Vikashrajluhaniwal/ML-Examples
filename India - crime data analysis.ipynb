#### Join/merge all these four data files pointing to four different crime types

#### Q21. Refer to the joined/merged data. Find out the total number of crimes in decreasing order for each state.

```python
df1 = distrCAW_modified.drop("Year",axis=1).groupby("STATE/UT").sum().sum(axis=1)
df2 = propertycrimes_modified.iloc[:,[0,3,5,7,9]].groupby("STATE/UT").sum().sum(axis=1)
df3 = murder_Crimes_modified.iloc[:,[0,3,4,5,6,7,8]].groupby("STATE/UT").sum().sum(axis=1)
df4 = KA_crimes[(KA_crimes["Pupose"]=="Total")==False].iloc[:,[0,3]].groupby("STATE/UT").sum().sum(axis=1)
df1.index = df1.index.str.lower()
df2.index = df2.index.str.lower()
df3.index = df3.index.str.lower()
df4.index = df4.index.str.lower()
df1.rename(index={'a & n islands':'a&n islands'}, inplace=True)
```

```python
final_joined = pd.concat([df1,df2,df3,df4], axis=1)
final_joined.sum(axis=1).sort_values(ascending = False)
```

#### Q22. Refer to the joined/merged data. Find out the total number of crimes in decreasing order for each year.

```python
df1 = distrCAW_modified.groupby("Year").sum().sum(axis=1)
df2 = propertycrimes_modified.iloc[:,[1,3,5,7,9]].groupby("YEAR").sum().sum(axis=1)
df3 = murder_Crimes_modified.iloc[:,[1,3,4,5,6,7,8]].groupby("YEAR").sum().sum(axis=1)
df4 = KA_crimes[(KA_crimes["Pupose"]=="Total")==False].iloc[:,[1,3]].groupby("YEAR").sum().sum(axis=1)
```

```python
final_joined_yearwise = pd.concat([df1,df2,df3,df4], axis=1)
final_joined_yearwise.sum(axis=1).sort_values(ascending = False)
```
